# Model

The **consumer** is described by:

1. A cobb-douglas utility function over leisure time ($f$) and consumption ($x$), $u(f,x)=f^{\alpha}x^{\beta}, \alpha,\beta >0$
2. A time endowment of $L>0$ and no endowment of the consumption good

The **firm** is described by

1. A production function, $y = g(\ell) = A\ell^{\gamma}, \gamma \in (0,1)$, where $\ell$ is labor input and $y$ is output of the consumption good
2. Profit is given by $\Pi = p \cdot y - w \cdot \ell = p \cdot g(\ell) - w \cdot \ell$


Both the **labor market** (wage $w$) and the **goods market** ($p$) is characterized by **perfect competition**. 

The **baseline parameters** are $\alpha = \beta = \gamma = A = 0.5$. The **numeraire** is $p=1$.

# Static

```python
# code
def walras(w):

    # a. consumer
    consumer = {}
    consumer['utility'] = 'cobb_douglas'
    consumer['alpha'] = 0.5
    consumer['beta'] = 0.5

    # b. firm
    firm = {}
    firm['gamma'] = 0.5
    firm['A'] = 0.5

    # c. model
    model = KoopmanModel(consumer,firm,w=w)
    model.walras_figure()
    
walras(0.2)
```

# Interactive: Auctioneer

```python
#code
widgets.interact(walras,w=widgets.BoundedFloatText(description='$w$',min=0.05, max=2.00, step=0.01, value=0.2));
```
