# Budget set

$$C(p_1,p_2,I) = \{(x_1,x_2) \in \mathbb{R}_{+}^2 \,\, | \,\, p_1 x_1 + p_2 x_2 \leq I\} $$

# Budget set with endowment

$$C(p_1,p_2,I) = \{(x_1,x_2) \in \mathbb{R}_{+}^2 \,\, | \,\, p_1 x_1 + p_2 x_2 \leq p_1 e_1 + p_2 e_2\} $$

```python
budgetset.draw_endowment();
```

# Kinked budget

$$ C(p_{1}^{A},p_{1}^{B},p_{2},I) =  \left\{ (x_{1},x_{2})\in\mathbb{R}_{+}^{2}\,|\,\begin{cases}
p_{1}^{A}x_{1}+p_{2}x_{2} \leq I & \text{hvis }x_{1}\leq\overline{x_{1}}\\
p_{1}^{B}(x_{1}-\overline{x})+p_{2}x_{2} \leq  I-p_{1}^{B}\overline{x} & \text{hvis }x_{1}>\overline{x}_{1}
\end{cases}\right\}  $$

```python
budgetset.draw_kink();
```

# Budget set with exogenous income in 3D

$$C(p_1,p_2,p_3,I) = \{(x_1,x_2,x_3) \in \mathbb{R}_{+}^3 \,\, | \,\, p_1 x_1 + p_2 x_2 + p_3 x_3 \leq I\} $$

```python
I = 10 # income
price_vectors = [] # list of price vectors
price_vectors.append([1,1,1]) # price vector (1,1,1)
price_vectors.append([2,1,1]) # price vector (2,1,1)
price_vectors.append([1,2,1]) # etc.
price_vectors.append([5,1,1])
price_vectors.append([1,5,1])
price_vectors.append([1,1,2])
price_vectors.append([1,1,5])
                     
# print
budgetset.draw_3D(price_vectors,I);
```
